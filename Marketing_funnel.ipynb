```python
Data='https://raw.githubusercontent.com/PALAK7890/Marketing_Funnel/refs/heads/main/marketing_funnel_dirty_60k.csv'
```

```python
import pandas as pd
import numpy as np

df=pd.read_csv(Data)
```

```python
df.info()
```

```python
df['device_type']=df['device_type'].fillna(df['device_type'].mode()[0])
```

```python
df['landing_page']=df['landing_page'].fillna(df['landing_page'].mode()[0])
```

```python
df['ad_spend_usd']=df['ad_spend_usd'].fillna(df['ad_spend_usd'].mean())
```

```python
df['country']=df['country'].fillna(df['country'].mode()[0])
```

```python
df['persona_segment']=df['persona_segment'].fillna(df['persona_segment'].mode()[0])
```

```python
df.info()
```

```python
df=df.drop_duplicates()
```

```python
df['device_type'] = df['device_type'].str.strip().str.lower()

df['device_type'] = df['device_type'].replace({
    'mobile': 'Mobile',
    'desktop': 'Desktop',
    'tab': 'Tablet',
    'tablet': 'Tablet'
})
```

```python
df['channel'] = df['channel'].str.strip().str.lower()

df['channel'] = df['channel'].replace({
    'google ads': 'Google Ads',
    'meta': 'Meta Ads',
    'linkedin ads': 'LinkedIn Ads',
    'organic': 'Organic Search',
    'youtube': 'YouTube'
})
```

```python
df['country'] = df['country'].str.strip()
```

```python
df = df[df['clicks'] <= df['impressions']]
df = df[df['signups'] <= df['clicks']]
df = df[df['activations'] <= df['signups']]
df = df[df['retained_users'] <= df['activations']]

df = df[df['ad_spend_usd'] >= 0]
```

```python
Q1 = df['ad_spend_usd'].quantile(0.25)
Q3 = df['ad_spend_usd'].quantile(0.75)
IQR = Q3 - Q1

df = df[(df['ad_spend_usd'] >= Q1 - 1.5*IQR) &
        (df['ad_spend_usd'] <= Q3 + 1.5*IQR)]
```

```python
df['event_date'] = pd.to_datetime(df['event_date'])
```

```python
df['CTR'] = df['clicks'] / df['impressions']
df['signup_rate'] = df['signups'] / df['clicks']
df['activation_rate'] = df['activations'] / df['signups']
df['retention_rate'] = df['retained_users'] / df['activations']
```

```python
df['quality_score'] = (
    0.3 * df['signups'] +
    0.3 * df['activations'] +
    0.4 * df['retained_users']
)
```

```python
df['signup_rate'] = df['signup_rate'].fillna(0)
df['activation_rate'] = df['activation_rate'].fillna(0)
df['retention_rate'] = df['retention_rate'].fillna(0)

df['month'] = df['event_date'].dt.to_period('M').astype(str)
```

```python
df.info()
```

```python
df.to_csv("marketing_funnel.csv", index=False)
```
